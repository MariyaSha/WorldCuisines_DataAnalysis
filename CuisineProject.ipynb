## 3 Web Scraping

### 3.1 Cuisine-Based Scraper Details

In this section, I will explain the logic behind each step of my scraper.
<br>
The implementation of this logic can be found in the code cell below.

- <b>SETP 1:</b> 
<br>
    selecting a desired list of cuisines and assigning it to "cuisines". This list can be adjusted to include many more categories.
    
```python
cuisines = ["Middle Eastern", "Southeast Asian", "Indian", "French", "Italian"]
```

- <b>SETP 2:</b> 
<br>

> Please note, if a different browser software is used, please adjust .Chrome() accordingly.

- <b>SETP 3:</b> 
<br>
    parsing over each cuisine and fetching all the recipes from the first 25 pages (500 reciper per cuisine).
    Each item in the list of chosen cuisine names, <i>cuisines</i>, is then concatinated into the URL string, which will navigate to each main cuisine page one at a time.

```python
for cuisine in cuisines:
    # navigate to RecipeDB and wait for the page to fully load
    driver.get("https://cosylab.iiitd.edu.in/recipedb/search_region/" + cuisine.replace(" ", "%20"))
```

- <b>SETP 5:</b>

    iterating over a long list of pages, fetching the recipe table from each page, converting it into a dataframe and adding a unique "primary_key" field. Lastly, inserting the fetched data into recipes.csv

``` python
# fetch data from the first n_pages
# n_pages can be any positive integer greated than 1
for page_n in range(1, n_pages):    
    # fetch the DOM table element (all the recipe names of the current page)
    recipe_table = driver.find_element(By.XPATH, "//table")
    # convert the table element into a dataframe
    recipe_df = pd.read_html("<table>" + recipe_table.get_attribute('innerHTML') + "</table>")    
    # add a "primary key" field to dataframe
    recipe_df[0]["primary_key"] = range(primary_key, primary_key + len(recipe_df[0]))    
    # insert dataframe into 'recipes.csv'
    recipe_df[0].to_csv('recipes.csv', mode='a', index=False, header=False)
```

- <b>SETP 6:</b> 
<br>
    fetch all the anchors that link to individual recipe pages and iterate over their value and indices using the enumerate() function:
    
 ``` python
# fetch links to individual recipe pages
recipe_links = driver.find_elements(By.XPATH,'.//a[contains(@href, "/recipedb/search_recipeInfo/")]')
# iterate over each recipe to fetch the ingredients
for idx, recipe in enumerate(recipe_links):
    recipe.click()
    time.sleep(5)
```
- <b>SETP 7:</b> 
<br>
    the above code would result in opening an individual recipe in another tab of the browser. 
    <br>
    To switch to the new tab, we will use the following command.
    
```python
driver.switch_to.window(driver.window_handles[1])
```

- <b>STEP 8:</b>
<br>
    in the new tab the highlighted section is <i>"preview recipe"</i>, while the section of interest is the <i>"ingredients"</i> section. So we will first navigate there using the following block of code:
    
```python
ingredients_section = driver.find_element(By.XPATH, "//a[@href='#ingredient_nutri']")
ingredients_section.click()
time.sleep(5)
```   

- <b>STEP 9:</b>
<br>
    once we are in the correct section of the page, we can then fetch the table element containing the ingredients and store it in ingredients.csv

```python
ingredients_table = driver.find_element(By.XPATH, "//table[@id='myTable']")
ingredients_df = pd.read_html("<table>" + ingredients_table.get_attribute('innerHTML') + "</table>")
ingredients_df[0]["recipe_key"] = primary_key
ingredients_df[0].to_csv('ingredients.csv', mode='a', index=False, header=False)
```  

- <b>STEP 10:</b>
<br>
    once the ingredients table was successfuly scraped - we will close the new tab and switch to the original tab of the recipe table. We will also increment the primary key for the next iteration:
    
```python
# close current tab
driver.close()
# switch to the original tab (list of recipes)
driver.switch_to.window(driver.window_handles[0])
primary_key += 1
```    

- <b>STEP 11:</b>
<br>
Once we are done scraping the entire list of recipes of a given page - we will navigate to the next page by clicking on the "Next" anchor:

```python
next_page = driver.find_element(By.XPATH,".//a[contains(text(), 'Next')]")
next_page.click()
```  

- <b>STEP 12:</b>
<br>
Repeat the above commands until all the desired cuisine pages and recipes were scraped. 
The entire process in 100% automated, no action is required from the user, only patience.

````python
# CUISINE BASED SCRAPER

cuisines = ["Middle Eastern", "Southeast Asian", "Indian", "French", "Italian"]

#initiallize a web browse object
#please note: chromedriver.exe file must be present in the root object
driver = webdriver.Chrome()

primary_key = 0

for cuisine in cuisines:
    # navigate to RecipeDB and wait for the page to fully load
    driver.get("https://cosylab.iiitd.edu.in/recipedb/search_region/" + cuisine.replace(" ", "%20"))
    time.sleep(5)   

    # fetch data from the first 25 pages (500 recipes)
    for page_n in range(25): 

        # fetch the DOM table element (all the recipe names of the current page)
        recipe_table = driver.find_element(By.XPATH, "//table")

        # convert the table element into a dataframe
        recipe_df = pd.read_html("<table>" + recipe_table.get_attribute('innerHTML') + "</table>")

        # add a "primary key" field to dataframe
        recipe_df[0]["primary_key"] = range(primary_key, primary_key + len(recipe_df[0]))

        # insert dataframe into 'recipes.csv'
        recipe_df[0].to_csv('recipes.csv', mode='a', index=False, header=False)

        # fetch links to individual recipe pages
        recipe_links = driver.find_elements(By.XPATH,'.//a[contains(@href, "/recipedb/search_recipeInfo/")]')

        # iterate over each recipe to fetch the ingredients
        for idx, recipe in enumerate(recipe_links):  

            # click on the recipe link          
            recipe.click()
            time.sleep(5)

            # navigate to the new tab that the previous block of code opened
            driver.switch_to.window(driver.window_handles[1])

            # select the ingredients section
            ingredients_section = driver.find_element(By.XPATH, "//a[@href='#ingredient_nutri']")
            ingredients_section.click()
            time.sleep(5)

            # fetch the list of ingredients and save it to 'ingredients.csv'
            ingredients_table = driver.find_element(By.XPATH, "//table[@id='myTable']")
            ingredients_df = pd.read_html("<table>" + ingredients_table.get_attribute('innerHTML') + "</table>")
            ingredients_df[0]["recipe_key"] = primary_key
            ingredients_df[0].to_csv('ingredients.csv', mode='a', index=False, header=False)

            # print output while scraping
            print("fetched:", primary_key)

            # close current tab
            driver.close()
            # switch to the original tab (list of recipes)
            driver.switch_to.window(driver.window_handles[0])

            # reset the primary key for the next iteration
            primary_key += 1

        time.sleep(2)

        # navigate to the next page of recipes
        next_page = driver.find_element(By.XPATH, ".//a[contains(text(), 'Next')]")
        next_page.click()

        time.sleep(2)

print("SCRAPER IS DONE! HAVE A NICE DAY!")       
````

### 5.1 Total Visualised Outcome

I will combine all the data we obtained above into one visual plot:

````python
proteins = [proteins_middle, proteins_south_asian, proteins_indian, proteins_french, proteins_italian]
fats = [fat_middle, fat_south_asian, fat_indian, fat_french, fat_italian]
index = ['Middle Eastern', 'Southeast Asian', 'Indian',
         'French', 'Italian']
df = pd.DataFrame({'proteins': proteins,
                   'fats': fats}, index=index)
ax = df.plot.bar(rot=90, title="nutritional profiles of different cuisines")
````

### 5.2 Nutritional Profile Conclusions

In the above plot, we can see that each cuisine has a different balance between fats and proteins.
- it seems that the French and the Indian cuisins have similar protein to fat ratios.
- it seems that the Itallian cuisine has almost an even balance between proteins and fats.
- it seems that the Middle Eastern cuisine is the only one that includes more proteins than fats.
- it seems that the Southeast Asien cuisine has the biggest gap between proteins and fats.

### 5.3 Gaps in my approach to nutritional profiling

The evaluation of the above nutritional values is not as accurate as it could be.
As each recipe has a different number and convention of <i>"Servings"</i> - it is difficult to determine the exact nutritional values.
<br>
In some recipes, <i>"Servings"</i> refers to a number of cookies or chocolate balls, in other recipes it relates to the number of main dishes or loafs.
The servings are not measured universaly, which makes the results approximate rather than obsolute.

### 5.4 Insights

- My analysis shows that each cuisine has a unique balance of nutritional values and it can be expended further introducing a larger number of samples. 
- In addition, even though many ingredients are universaly popular - among the top 10 most used ingredients in each cuisine we can find unique values.
<br>
For example: butter is the top universal ingredient, however - it doesn't appear in the top 10 of the Middle Eastern, Southeast Asian and Indian cuisines.
<br>
- Some cuisines relate to one another. 
<br>
For example: 6 of the top 10 ingredients of the French and Itallian cuisines are a perfect match! In particular: olive oil, salt, butter, onion, black pepper, sugar
